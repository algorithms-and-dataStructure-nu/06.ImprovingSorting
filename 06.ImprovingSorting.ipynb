## Merge Sort
Merge Sort (マージソート) は、問題を解決可能な小さな問題に分割し、その結果を統合することで解決する。こうした方法を、「分割統治法 (Divide and Conquer)」と呼ぶ。
- 整列済みの２つのリストをマージ (結合) して、１つの整列済みリストを作成する手順
    - ２つのリストの先頭要素を比較し、小さい方を結果リストに追加する
    - ２つのリストの長さの和程度の比較で十分
- もとのリストを再帰的に半分に分割し、長さ１のリストにする
    - 長さ１のリストは整列済み
- 分割したリストをマージしていく

----
The Merge Sort algorithm divides the problem into smaller subproblems that are easier to solve, and then combines the results to solve the original problem. This approach is known as "Divide and Conquer."
- Merging two sorted lists into one sorted list:
    - Compare the first elements of both lists and add the smaller one to the result list.
    - This process requires a number of comparisons proportional to the sum of the lengths of the two lists.
- Recursively divide the original list in half until each sublist has a length of one.
    - A list of length one is already sorted.
- Merge the divided lists back together.
----

### Merge sorted two lists

```python
def mergeList(left:list, right:list) -> list:
    """
    左と右の整列済みリストをマージして一つの整列済みリストにする
    
    Merge two sorted lists into a single sorted list.
    """
    merged = []
    i = 0
    j = 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            merged.append(left[i])
            i += 1
        else:
            merged.append(right[j])
            j += 1
    # どちらかのリストが空になったら残りを追加
    # Append the remaining elements of either list if one is exhausted
    merged.extend(left[i:])
    merged.extend(right[j:])
    return merged
```

### Merge Sort
Merge Sortは、以下のような再帰的な手順で実装できる。
- リストが長さ1以下の場合は、そのリストは整列済みであり、そのまま返す
- リストを半分に分割し、左右それぞれをMerge Sortで整列する
- 整列済みの左右のリストを`mergeList()`でマージして返す
----
Merge Sort can be implemented using the following recursive procedure:
- If the list has a length of 1 or less, it is already sorted; return it as is.
- Split the list in half and sort both halves using Merge Sort.
- Merge the two sorted lists using `mergeList()` and return the result.


```python
def mergeSort(arr:list) -> list:
    """
    マージソートアルゴリズムでリストを整列する
    
    Sort a list using the merge sort algorithm.
    """
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = mergeSort(arr[:mid])
    right = mergeSort(arr[mid:])
    return mergeList(left, right)
```

```python
data = [38, 27, 43, 3, 9, 82, 10, 55]
print("Original list:", data)
sorted_data = mergeSort(data)
print("Sorted list:", sorted_data)
```

## クィックソート: Quick Sort
クィックソート (Quick Sort) は、基準値 (ピボット) を選び、基準値より小さい要素のリストと大きい要素のリストに分割し、再帰的に整列する。
- ここでは、各部分リストの最後尾の要素を基準値とする
- 基準値より小さい要素のリスト、基準値と同じ要素のリスト、基準値以上の要素のリストを作成
- 再帰的に各部分リストを整列
- 最後に、整列済みの小さい要素のリスト、基準値と同じ要素のリスト、整列済みの大きい要素のリストを結合
----
Quick Sort is a sorting algorithm that selects a pivot value, partitions the list into elements less than the pivot and elements greater than or equal to the pivot, and recursively sorts the sublists.
- In this implementation, the last element of each sublist is chosen as the pivot.
- Create a list of elements less than the pivot and a list of elements greater than or equal to the pivot.
- Recursively sort each sublist.

```python
def quickSort(arr:list) -> list:
    """
    クイックソートアルゴリズムでリストを整列する
    
    Sort a list using the quick sort algorithm.
    """
    if arr is None or len(arr) <= 1:
        return arr
    pivot = arr[-1]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    left = quickSort(left)
    right = quickSort(right)
    return left + middle + right
```

```python
data = [38, 27, 43, 3, 9, 82, 10, 55, 27, 3, 82, 10, 55, 100, 1, 4, 530]
print("Original list:", data)
sorted_data = quickSort(data)
print("Sorted list:", sorted_data)
```
